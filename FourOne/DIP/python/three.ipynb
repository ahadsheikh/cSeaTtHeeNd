```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Images resourses..
stadium = cv.imread('../assets/stadium.tif')
stadium = cv.cvtColor(stadium, cv.COLOR_BGR2GRAY)

noisy_image = cv.imread('../assets/noisy_image.tif')
noisy_image = cv.cvtColor(noisy_image, cv.COLOR_BGR2GRAY)
plt.imshow(noisy_image, cmap='gray')
```

```python
# Qiestion number 3
mask_size = 3
m, n = noisy_image.shape
mask = np.ones([mask_size, mask_size], dtype=int)
mask = mask/9
new_image = np.zeros([m, n])

for i in range(1, m-1):
    for j in range(1, n-1):
        temp = noisy_image[i-1, j-1]*mask[0, 0]+noisy_image[i-1, j]*mask[0, 1]+noisy_image[i-1, j + 1]*mask[0, 2]+noisy_image[i, j-1]*mask[1, 0]+ noisy_image[i, j]*mask[1, 1]+new_image[i, j + 1]*mask[1, 2]+noisy_image[i + 1, j-1]*mask[2, 0]+noisy_image[i + 1, j]*mask[2, 1]+noisy_image[i + 1, j + 1]*mask[2, 2]
        new_image[i, j]= temp

new_image = new_image.astype(np.uint8)
cv.imshow(new_image, cmap='gray')
```
